# BA1F: Find a Position in a Genome Minimizing the Skew

```python
import matplotlib.pyplot as plt

def min_skew_position(genome):
    min_skew = float('inf')
    skew = 0
    min_positions = []

    skew_values = []  # Store skew values for plotting

    for i, nucleotide in enumerate(genome):
        if nucleotide == 'G':
            skew += 1
        elif nucleotide == 'C':
            skew -= 1

        skew_values.append(skew)  # Append skew value for plotting

        if skew < min_skew:
            min_skew = skew
            min_positions = [i + 1]  # Add 1 to convert from 0-based index to 1-based index
        elif skew == min_skew:
            min_positions.append(i + 1)

    return min_positions, skew_values

sequence = "CCTATCGGTGGATTAGCATGTCCCTGTACGTTTCGCCGCGAACTAGTTCACACGGCTTGATGGCAAATGGTTTTTCCGGCGACCGTAATCGTCCACCGAG"

min_positions, skew_values = min_skew_position(sequence)

# Plot the skew graph
plt.figure(figsize=(12, 6))
plt.plot(range(len(sequence)), skew_values)
plt.xlabel("Position in Sequence")
plt.ylabel("Skew")
plt.title("Skew Graph for the Sequence")
plt.grid(True)
plt.savefig("skew_graph.png")  # Save the plot to a file (optional)
plt.xticks(range(len(sequence)+1),list(sequence))
plt.show()


print(" ".join(map(str,min_positions)))
```
